```python
import os
import keras
import numpy as np
import pandas as pd
#import cnn_utils as utils
#from keras.utils import np_utils
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.image import imread
from PIL import Image
from keras.utils import plot_model
from keras.models import load_model
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
```

## Tratamento das imagens para predição
### Usando PIL

```python
# Carregando modelo
leaf_model = load_model('CassavaLeafDisease.h5')
# Imagem teste
imagem_teste = 'dados/test_images/2216849948.jpg'
```

```python
# Carrefando imagem 
teste1 = Image.open(imagem_teste)
teste1
```

```python
# Redimencionado imagem  800x600 para 456x456
teste1 = teste1.resize((456, 456), resample = Image.NEAREST) 
teste1
```

```python
# Trasformando e redimencionando para ndarray (1, 456, 456, 3)
teste1 = np.expand_dims(teste1, axis = 0)
teste1
```

```python
teste1.shape
```

```python
prediction1 = leaf_model.predict(teste1)
prediction1
```

### Usando keras

```python
# Carrefando imagem e redimencionando de 800x600 para 456x456
teste2 = image.load_img(imagem_teste, target_size = (456, 456))
teste2
```

```python
# Trasformando para ndarray (456, 456, 3)
teste2 = image.img_to_array(teste2)
teste2
```

```python
teste2.shape
```

```python
# Redimencionando para ndarray (1, 456, 456, 3)
teste2 = np.expand_dims(teste2, axis = 0)
teste2
```

```python
teste2.shape
```

```python
prediction2 = leaf_model.predict(teste2)
prediction2
```
